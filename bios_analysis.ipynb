```python
import pandas as pd
```

```python
bios=pd.read_csv('./bios.csv') # .read_csv() reads csv files
bios.head()
```

```python
# we will copy the bios to a new dataframe and do cleaning on that and keep the bios as it is
df=bios.copy()
```

```python
# replacing the • in used names to a single empty space " "
df['name']=df['Used name'].str.replace("•"," ") # this will add a new column buw without the • in names
```

```python
df.head()
```

```python
# Split the Measurements Column into height and weight Column

df[['Height_cm','Weight_kg']]=df['Measurements'].str.split("/",expand=True)
# expand=True parameter splits the column into 2 different columns 
```

```python
df.head()
```

```python
# we also need to get rid of cm and kg in our heifht_cm and weight_kg column

df["Height_cm"]=pd.to_numeric(df['Height_cm'].str.strip(' cm'),errors='coerce')
df["Weight_kg"]=pd.to_numeric(df['Weight_kg'].str.strip(' kg'),errors='coerce')
df
```

```python
# get the tallest player
df.sort_values('Height_cm',ascending=False)
```

```python
# Parse out Dates from Born Column
# we use regex here
date_pattern=r'(\d+ \w+ \d{4}|\d{4})'
df['born_date']=df['Born'].str.extract(date_pattern)

# converting born_date to datetime dtype
df['born_date']=pd.to_datetime(df['born_date'],format='mixed',errors='coerce')
# new column born_date added
df
```

```python
# we still need to check if there are other patterns in born column
df[~df['Born'].str.match(date_pattern,na=False) & df["Born"].notna()]
# there are indeed different dates some are only year 
# here we have to be the judge whether we have to add this king of data or let go of it
# ideally we should capture more info as possible
```

```python
df.info()
# check born_date dtype
```

```python
# sorting values based on born_date
df.sort_values('born_date',ascending=False)
```

```python
# Parse out Dates from Died Column
# code is exactly similar to cell 15 compare cell 15 code with this code
# we use regex here
date_pattern=r'(\d+ \w+ \d{4}|\d{4})'
df['Died_date']=df['Died'].str.extract(date_pattern)

# converting Died_date to datetime dtype
df['Died_date']=pd.to_datetime(df['Died_date'],format='mixed',errors='coerce')
# new column Died_date added
df
```

```python
df.info()
```

```python
# parsing out city, region and country from Born column
location_pattern=r'in ([\w\s-]+), ([\w\s-]+) \((\w+)\)'
```

```python
df[['born_city','born_region','born_country']]=df['Born'].str.extract(location_pattern,expand=True)
df
```

```python
# diaplaying where most olympic athletes are born (first 10)
df.value_counts('born_region',ascending=False).head(10)
```

```python
# converting the dtype of each column
# changed the dtype of columns
df['athlete_id']=df['athlete_id'].astype('int')
df['name']=df['name'].astype('string')
```

```python
df['born_city']=df['born_city'].astype('string')
df['born_region']=df['born_region'].astype('string')
```

```python
df['born_country']=df['born_country'].astype('string')
df['NOC']=df['NOC'].astype('string')
```

```python
df['Height_cm']=df['Height_cm'].astype('int32',errors="ignore")
df['Weight_kg']=df['Weight_kg'].astype('int32',errors='ignore')
```

```python
df.info()
```

```python
# choosing which columns to keep
columns_to_keep=['athlete_id','name','born_date','born_city','born_region',
                 'born_country','NOC','Height_cm','Weight_kg',"Died_date"]
df_clean=df[columns_to_keep]

# alternatively i think that we should drop the columns which are 
# not usefulsuch as 'Born' Column  then 'Used name' is another such column
# all in all you have to decide for yourself what small cleaning you can do 
# to the original data and try to keep maximum amount of data in clean table

# more cleaning can also be done at this point its just a matter of choice
```
